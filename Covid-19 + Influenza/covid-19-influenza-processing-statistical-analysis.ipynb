```python
def rename_influenza_columns_pt(data):
    
    new_names = ['Gênero','Febre','Tosse','Dor de Garganta','Dispnéia',
                 'Desconforto Respiratório','Saturação do Oxigenio','Diarreia',
                 'Vomito','Dor Abdominal','Fadiga', 'Distúrbios do Olfato',
                 'Distúrbio do paladar','Classe']
    
    data.columns = new_names
    data['Classe'] = data['Classe'].replace(0,'pi').replace(1,'ni')
    return data

def rename_covid_columns_pt(data):
    
    new_names = ['Dor de Garganta','Dispnéia','Febre','Tosse','Dor de Cabeça',
                 'Distúrbio do paladar','Distúrbios do Olfato','Coriza','Gênero',
                 'Profissional de Saúde?','Classe']
    
    data.columns = new_names
    data['Classe'] = data['Classe'].replace(0,'pc').replace(1,'nc')
    return data

def read_covid_dataset(file_name):
    df = pd.read_csv(file_name)
    df = df.astype(int)
    return rename_covid_columns_pt(df)

def read_influenza_dataset(file_name):
    df = pd.read_csv(file_name)
    del df['Unnamed: 0']
    df = df.astype(int)
    df = df.replace(0,'n').replace(1,'p').replace('n',1).replace('p',0).replace(2.0,1)
    return rename_influenza_columns_pt(df)

def remove_columns(df, removed_col):
   
    for col in removed_col:
        del df[col]
    return df


```

# Merge covid-19 + influenza datasets

```python
def merge_data(dt1,dt2,new_dt_name, mach_class=True):
    df_merged = pd.concat([dt1,dt2])
    df_merged = df_merged.fillna(1)
    df_merged['Classe'] = df_merged['Classe'].replace('pc',0).replace('nc',1).replace('pi',2).replace('ni',3)
    df_merged = df_merged.astype(int)

    df_merged.reset_index(drop=True, inplace=True)
    if mach_class:
         df_merged = match_classes(df_merged)
    
    else:
        classes = df_merged['Classe']
        del df_merged['Classe']
        df_merged['Classe'] = classes
        
    print(df_merged['Classe'].value_counts())
    df_merged.to_csv(new_dt_name)
    return df_merged

def match_classes(dt):
    y=dt["Classe"]
    del dt["Classe"]
    x=dt.iloc[:,:]
    x['index']=dt.index
    nm = NearMiss()
    x,y=nm.fit_resample(x,y)
    x['Classe'] = y
    del x['index']
    return x
```

```python
# #balanced
df_rapid_covid_inf = merge_data(df_rapid_covid,df_rapid_influ,'df_rapid_covid_inf.csv')
df_pcr_covid_inf = merge_data(df_pcr_covid,df_pcr_influ,'df_pcr_covid_inf.csv')
df_both_covid_inf = merge_data(df_both_covid,df_both_influ,'df_both_covid_inf.csv')

# #unbalanced
df_rapid_unb_covid_inf = merge_data(df_rapid_unb_covid,df_rapid_unb_influ, 'df_rapid_covid_inf_unb.csv',mach_class=False)
df_pcr_unb_covid_inf = merge_data(df_pcr_unb_covid,df_pcr_unb_influ,'df_pcr_covid_inf_unb.csv',mach_class=False)
df_both_unb_covid_inf = merge_data(df_both_unb_covid,df_both_unb_influ,'df_both_covid_inf_unb.csv',mach_class=False)
```

# Statistical Analysis

```python
def set_crosstab(dt):
   
    final_crosstab = []
    for c in dt.columns[:16]:
        data=pd.crosstab(index=dt[c],columns=dt["Classe"])
        data['sum']=None
        data['p_t']=None
        data['p_p']=None
        data['p_n']=None
        data.iloc[:1,4:5]=data[0][0]+data[1][0]+data[2][0]+data[3][0]
        data.iloc[1:2,4:5]=data[0][1]+data[1][1]+data[2][1]+data[3][1]
        data.iloc[:1,5:6]=data.iloc[:1,4:5].values[0][0]*100/len(dt)
        data.iloc[1:2,5:6]=data.iloc[1:2,4:5].values[0][0]*100/len(dt)
        data.iloc[:1,6:7]=((data[0][0] + data[2][0])/data.iloc[:1,4:5].values[0][0]) * data.iloc[:1,5:6].values[0][0]
        data.iloc[1:2,6:7]=((data[0][1] + data[2][1])/data.iloc[1:2,4:5].values[0][0]) * data.iloc[1:2,5:6].values[0][0]
        data.iloc[:1,7:8] = data.iloc[:1,5:6].values[0][0] - data.iloc[:1,6:7].values[0][0]
        data.iloc[1:2,7:8]= data.iloc[1:2,5:6].values[0][0] - data.iloc[1:2,6:7].values[0][0]
        print(data, '\n')
        
def test_statistic_chi2(value,data):
    test_result=pd.crosstab(index=data[value],columns=data["Classe"])
    test_statistic, p_value, dof, expected=stats.chi2_contingency(test_result.iloc[:,:].values,correction=True)
    alpha = 0.01
    if p_value <= alpha:
        print('Dependent (reject H0)')
    else:
        print('Independent (fail to reject H0)')
        
def print_stattistic_chi2_result(data):
    for column in data.columns[:16]:
        print(f'{column} :')
        test_statistic_chi2(column,data)
        print('\n')
        
```
