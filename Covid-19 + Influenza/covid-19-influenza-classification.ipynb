```python
def RandomForest_classif(x_train,y_train,param):
    #Classification
    
    
    clf= RandomForestClassifier(**param)
    clf=clf.fit(x_train, y_train)
    filename = 'randomForest_model.sav'
    pickle.dump(clf, open(filename, 'wb'))
    #filename.to_csv('randomForest_model.sav',index=False)
    return clf

def Kneighbors_classif(x_train,y_train,param):
    #Classification
    

    clf= KNeighborsClassifier(**param)
    clf= clf.fit(x_train, y_train)
    filename = 'kneighbors_model.sav'
    pickle.dump(clf, open(filename, 'wb'))
    #filename.to_csv('kneighbors_model.sav',index=False)
    return clf

def DecisionTree_classif(x_train,y_train,param):
    #Classification
    
    clf = tree.DecisionTreeClassifier(**param)
    clf = clf.fit(x_train,y_train)
    filename = 'decisionTree_model.sav'
    pickle.dump(clf, open(filename, 'wb'))
    #filename.to_csv('decisionTree_model.sav',index=False)
    return clf

def mpl_classif(x_train,y_train,param):
    
    clf =  MLPClassifier(**param)
    clf=clf.fit(x_train,y_train)
    filename = 'mlp_model.sav'
    pickle.dump(clf, open(filename, 'wb'))
    #filename.to_csv('mlp_model.sav',index=False)
    return clf

def gb_classif(x_train,y_train,param):
    
    clf=GradientBoostingClassifier(**param)
    clf=clf.fit(x_train, y_train)
    filename = 'gradientBoosting_model.sav'
    pickle.dump(clf, open(filename, 'wb'))
    #filename.to_csv('gradientBoosting_model.sav',index=False)
    return clf

def xgb_classif(x_train,y_train,param):
    clf = xgb.XGBClassifier(**param)
    filename = 'xgb_model.sav'
    pickle.dump(clf, open(filename, 'wb'))
    #filename.to_csv('xbg_model.sav',index=False)
    return clf.fit(x_train, y_train)

def svc_classif(x_train,y_train,param):
    
    regr = svm.SVC(**param)
    regr=regr.fit(x_train, y_train)
    filename = 'svc_model.sav'
    pickle.dump(regr, open(filename, 'wb'))
    #filename.to_csv('svc_model.sav',index=False)
    return regr

def lg_with_regu_classif(x_train,y_train,param):
    
    clf = LogisticRegression(** param)
    clf=clf.fit(x_train,y_train)
    filename = 'lg_regu_model.sav'
    pickle.dump(clf, open(filename, 'wb'))
    #filename.to_csv('lg_regu_model.sav',index=False)
    return clf

def lg_without_regu_classif(x_train,y_train,param):
    
    clf = LogisticRegression(** param)
    clf=clf.fit(x_train,y_train)
    filename = 'lg_model.sav'  
    pickle.dump(clf, open(filename, 'wb'))
    #filename.to_csv('lg_model.sav',index=False)
    return clf

```

```python
from itertools import cycle
def print_roc_curve(fpr,tpr,roc_auc,title,fig_path,n_classes,macro_roc):
    
    colors = cycle(['blue', 'red', 'green','yellow'])
    plt.plot(
    macro_roc["fpr"],
    macro_roc["tpr"],
    label="curva ROC média-macro (AUC = {0:0.2f})".format(macro_roc["auc"]),
    color="deeppink",
    linestyle=":",
    linewidth=4)
    for i, color in zip(range(n_classes), colors):
        plt.plot(fpr[i], tpr[i], color=color,
                 label='Classe {0} (AUC = {1:0.2f})'
                 ''.format(i, roc_auc[i]))

    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([-0.05, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('Taxa de Falso Positivo')
    plt.ylabel('Taxa de Verdadeiro Positivo')
    plt.title(title)
    plt.legend(loc="lower right")
    plt.savefig(fig_path)
#     plt.close()
    plt.show()

    
def print_prc_curve(recall, precision,ap, title,fig_path,n_classes):
    
    colors = cycle(['blue', 'red', 'green','yellow'])
    
    plt.plot(
    recall['macro'],
    precision['macro'],
    label="curva PR média-macro (AP = {0:0.2f})".format(ap['macro']),
    color="deeppink",
    linestyle=":",
    linewidth=4)
              
    for i, color in zip(range(n_classes), colors):
        
        plt.plot(recall[i], precision[i], color=color,
                 label='Classe {0} (AP = {1:0.2f})'
                 ''.format(i, ap[i]))

    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([-0.05, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('Recall')
    plt.ylabel('Precisão')
    plt.title(title)
    plt.legend(loc="lower right")
    plt.savefig(fig_path.replace('roc', 'pr'))
#     plt.close()
    plt.show()
```

```python
def calculate_metrics(x_train, x_test,y_train, y_test,data,k,clf,lista,p,r,path_name,title):
        
        from sklearn.metrics import recall_score, precision_score
        from itertools import cycle
        
        labels=[0,1,2,3]
        #prediction
        y_pred=clf.predict(x_test)
        
#         accuracy score
        acc=accuracy_score(y_test,y_pred)*100
        
        if k in [9,19,29,39,49]:
            y_test_2, y_pred_2 = y_test, y_pred
            y_test_2 = label_binarize(y_test_2.values, classes=labels)
            y_pred_2 = label_binarize(y_pred_2, classes=labels)
            n_classes = y_test_2.shape[1]
        
            fpr = dict()
            tpr = dict()
            roc_auc = dict()
            macro_roc = dict()

            for i in range(n_classes):
                fpr[i], tpr[i], _ = metrics.roc_curve(y_test_2[:, i], y_pred_2[:, i])
                roc_auc[i] = metrics.auc(fpr[i], tpr[i])

            all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))

            # Then interpolate all ROC curves at this points
            mean_tpr = np.zeros_like(all_fpr)
            for i in range(n_classes):
                mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])

            # Finally average it and compute AUC
            mean_tpr /= n_classes
            macro_roc["tpr"] = mean_tpr
            macro_roc["fpr"] = all_fpr
            macro_roc["auc"]  =  metrics.auc(all_fpr, mean_tpr)

            print_roc_curve(fpr,tpr,roc_auc,title,path_name,n_classes,macro_roc)


          # precision recall curve
            precision = dict()
            recall = dict()
            ap = dict()
            
            for i in range(n_classes):
                precision[i], recall[i], _ = precision_recall_curve(y_test_2[:, i],
                                                                    y_pred_2[:, i])
                ap[i] = metrics.average_precision_score(y_test_2[:, i],y_pred_2[:, i])
                
            # Compute micro-average ROC curve and ROC area
            precision["macro"], recall["macro"], _ = precision_recall_curve(y_test_2.ravel(),
             y_pred_2.ravel())
            ap["macro"] = metrics.average_precision_score(y_test_2, y_pred_2,
                                                     average="macro")
            
            print_prc_curve(recall, precision,ap,title,path_name,n_classes)
        
        #precision
        data.iloc[k:k+1,:1]= precision_score(y_test,y_pred, average='macro')
        #acc
        data.iloc[k:k+1,1:2]=acc
        
        #recall
        
        data.iloc[k:k+1,2:3] = recall_score(y_test,y_pred, average='macro')
        #mean absolute error
        data.iloc[k:k+1,3:4]=mean_absolute_error(y_test, y_pred)*100
        
        #AUC
        
        y_pred = label_binarize(y_pred, classes=labels)
        y_test = label_binarize(y_test.values, classes=labels)
     
        data.iloc[k:k+1,4:5] =  metrics.roc_auc_score(y_test,y_pred)*100
        # Brier score
        data.iloc[k:k+1,5:6]=brier_score_loss(y_test.reshape(-1,1),y_pred.reshape(-1,1))
        
        return data,lista,p,r
    
def calculate_metrics_voting(x_train, x_test,y_train, y_test,data,k,clf):
        
        
        #prediction
        y_pred=clf.predict(x_test)
        labels=[0,1,2,3]
        #accuracy score
        acc=accuracy_score(y_test,y_pred)*100
        
        
        #precision
        data.iloc[k:k+1,:1]= precision_score(y_test,y_pred, average='macro')
        
        data.iloc[k:k+1,1:2]=acc
        
        #recall
        data.iloc[k:k+1,2:3]= recall_score(y_test,y_pred, average='macro')
        
        
        #mean absolute error
        data.iloc[k:k+1,3:4]=mean_absolute_error(y_test, y_pred)*100
        
       #AUC
        y_pred = label_binarize(y_pred, classes=labels)
        y_test = label_binarize(y_test.values, classes=labels)
     
        data.iloc[k:k+1,4:5] =  metrics.roc_auc_score(y_test,y_pred)*100
        # Brier score
        data.iloc[k:k+1,5:6]=brier_score_loss(y_test.reshape(-1,1),y_pred.reshape(-1,1))
        
        
        
        return data
```

```python
def k_fold_cross_validation(x,y,lista_param,path_name,title):
    lista1, lista2, lista3, lista4,lista5,lista6,lista7,lista8,lista9 = ([] for i in range(9))
    p1, p2, p3, p4,p5,p6,p7,p8,p9 = ([] for i in range(9))
    r1, r2, r3, r4,r5,r6,r7,r8,r9 = ([] for i in range(9))
    dt=pd.DataFrame(columns=['P','ACC','R','MAE','AUC','BS','tn', 'fp', 'fn', 'tp'],index=range(50))
    rf=pd.DataFrame(columns=['P','ACC','R','MAE','AUC','BS'],index=range(50))
    dmlp=pd.DataFrame(columns=['P','ACC','R','MAE','AUC','BS'],index=range(50))
    dgbm=pd.DataFrame(columns=['P','ACC','R','MAE','AUC','BS'],index=range(50))
    xgboost=pd.DataFrame(columns=['P','ACC','R','MAE','AUC','BS'],index=range(50))
    dt_svm=pd.DataFrame(columns=['P','ACC','R','MAE','AUC','BS'],index=range(50))
    dt_knn=pd.DataFrame(columns=['P','ACC','R','MAE','AUC','BS'],index=range(50))
    dt_lg_regu=pd.DataFrame(columns=['P','ACC','R','MAE','AUC','BS'],index=range(50))
    dt_lg=pd.DataFrame(columns=['P','ACC','R','MAE','AUC','BS'],index=range(50))
                        
    kf=RepeatedStratifiedKFold(n_splits=10, n_repeats=5, random_state=42)
                        
    
    kf.get_n_splits(x,y)
    k=0
    for train_index, test_index in kf.split(x,y):

        x_train, x_test = x.iloc[train_index], x.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        clf_dt=DecisionTree_classif(x_train,y_train,lista_param[0])
        al_path = f'{path_name}/roc_curve-{k}-dt.jpg'
        dt,lista1,p1,r1=calculate_metrics(x_train, x_test,y_train, y_test,dt,k,clf_dt,
                                          lista1,p1,r1,al_path,title)

        clf_rf=RandomForest_classif(x_train,y_train,lista_param[1])
        al_path = f'{path_name}/roc_curve-{k}-rf.jpg'
        rf,lista2,p2,r2=calculate_metrics(x_train, x_test,y_train, y_test,rf,k,clf_rf,
                                          lista2,p2,r2,al_path,title)

        clf_svm=svc_classif(x_train,y_train,lista_param[2])
        al_path = f'{path_name}/roc_curve-{k}-svm.jpg'
        dt_svm,lista3,p3,r3=calculate_metrics(x_train, x_test,y_train, y_test, dt_svm,k,clf_svm,
                                              lista3,p3,r3,al_path,title)

        clf_knn=Kneighbors_classif(x_train,y_train,lista_param[3])
        al_path = f'{path_name}/roc_curve-{k}-knn.jpg'
        dt_knn,lista4,p4,r4=calculate_metrics(x_train, x_test,y_train, y_test, dt_knn,k,clf_knn,
                                              lista4,p4,r4,al_path,title)

        clf_gbm=gb_classif(x_train,y_train,lista_param[4])
        al_path = f'{path_name}/roc_curve-{k}-gbm.jpg'
        dgbm,lista5,p5,r5=calculate_metrics(x_train, x_test,y_train, y_test,dgbm,k,clf_gbm,
                                            lista5,p5,r5,al_path,title)

        clf_lg_regu=lg_with_regu_classif(x_train,y_train,lista_param[5])
        al_path = f'{path_name}/roc_curve-{k}-lgregu.jpg'
        dt_lg_regu,lista6,p6,r6=calculate_metrics(x_train, x_test,y_train, y_test,dt_lg_regu,k,
                                                  clf_lg_regu,lista6,p6,r6,al_path,title)

        clf_lg=lg_without_regu_classif(x_train,y_train,lista_param[6])
        al_path = f'{path_name}/roc_curve-{k}-lg.jpg'
        dt_lg,lista7,p7,r7=calculate_metrics(x_train, x_test,y_train, y_test,dt_lg,
                                             k,clf_lg,lista7,p7,r7,al_path,title)

        clf_mlp=mpl_classif(x_train,y_train,lista_param[7])
        al_path = f'{path_name}/roc_curve-{k}-mlp.jpg'
        dmlp,lista8,p8,r8=calculate_metrics(x_train, x_test,y_train, y_test,dmlp,k,clf_mlp,
                                            lista8,p8,r8,al_path,title)

        clf_xgboost=xgb_classif(x_train,y_train,lista_param[8])
        al_path = f'{path_name}/roc_curve-{k}-xgboost.jpg'
        xgboost,lista9,p9,r9=calculate_metrics(x_train, x_test,y_train, y_test,xgboost,k,clf_xgboost,
                                               lista9,p9,r9,al_path,title)
        print(k)
        k+=1
    data=pd.DataFrame([dt.iloc[:,:6].mean(),rf.iloc[:,:6].mean(),dgbm.iloc[:,:6].mean(),
                       xgboost.iloc[:,:6].mean(),dmlp.iloc[:,:6].mean(),dt_svm.iloc[:,:6].mean(),
                       dt_knn.iloc[:,:6].mean(),dt_lg_regu.iloc[:,:6].mean(),dt_lg.iloc[:,:6].mean()])


    return  data
```

```python
def majority_voting(x,y,estimator):
    
    dt=pd.DataFrame(columns=['P','ACC','R','MAE','AUC','BS'],index=range(50))
    
    kf=RepeatedStratifiedKFold(n_splits=10, n_repeats=5, random_state=42)
                        
    
    kf.get_n_splits(x,y)
    k=0
    for train_index, test_index in kf.split(x,y):

        x_train, x_test = x.iloc[train_index], x.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        
        # Voting Classifier with hard voting 
        vot_hard = VotingClassifier(estimators = estimator, voting ='hard') 
        vot_hard=vot_hard.fit(x_train, y_train) 
        
        dt= calculate_metrics_voting(x_train, x_test,y_train, y_test, dt,k,vot_hard)
        k+=1
    return dt.mean()
```
