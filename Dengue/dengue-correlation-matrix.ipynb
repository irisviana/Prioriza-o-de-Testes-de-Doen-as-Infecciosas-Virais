```python
both_data = pd.read_csv("../input/dengue-dataset/both_dengue_data.csv")

fem = pd.read_csv("../input/dengue-dataset/fem_dengue_data.csv")

mas = pd.read_csv("../input/dengue-dataset/ma_dengue_data.csv")


both_data_unb = pd.read_csv("../input/dataset-dengue-unbalanced/both_dengue_data_unb.csv")

fem_unb = pd.read_csv("../input/dataset-dengue-unbalanced/fem_dengue_data_unb.csv")

mas_unb = pd.read_csv("../input/dataset-dengue-unbalanced/ma_dengue_data_unb.csv")
```

```python
def rename_columns_index_pt(data):
    
    new_names = ['Gênero','Febre','Mialgia','Dor de Cabeça','Exantema',
                 'Vomito','Náusea','Dor nas costas',
                 'Conjuntivite','Artrite','Artralgia', 'Petéquia',
                 'Leucopenia','Dor no Retro','Classe']
    
    data.columns = new_names
    data.index = new_names
    return data
```

```python
def get_corr_matrix(data,lang='en'):
    cor_data = data.corr()
    cor_data = cor_data.round(2)
    if lang == 'pt':
        cor_data = rename_columns_index_pt(cor_data)
    else:
        cor_data = rename_columns_index_en(cor_data)
    return cor_data
```

```python
def plot_correlation_matrix(df_both,df_pcr,df_rapid,title_both,title_pcr,
                            title_rapid,color_type,fig_name, lang):
    
    cor_both_data = get_corr_matrix(df_both,lang)
    cor_pcr_data = get_corr_matrix(df_pcr,lang)
    cor_rapid_data = get_corr_matrix(df_rapid,lang)
    
    fig, ax =plt.subplots(1,3,figsize=(15,5))
    sns.heatmap(cor_both_data, annot=False,linewidths=.5, cmap=color_type, ax=ax[0])
    ax[0].set_title(title_both)
    sns.heatmap(cor_pcr_data, annot=False,linewidths=.5,cmap=color_type, ax=ax[1])
    ax[1].set_title(title_pcr)
    sns.heatmap(cor_rapid_data, annot=False,linewidths=.5, cmap=color_type, ax=ax[2])
    ax[2].set_title(title_rapid)
    fig.tight_layout(pad=0.5)
    plt.savefig(fig_name)
    fig.show()
```

```python
plot_correlation_matrix(both_data, fem, mas,
                        'Matriz de Correlação de Ambos os Genêros (Bal.)',
                        'Matriz de Correlação do Genêro Feminino (Bal.)',
                        'Matriz de Correlação do Genêro Masculino (Bal.)',
                        "RdPu",'dengue_correlation_matrix_pt_balanced.png', 'pt')
```

```python
plot_correlation_matrix(both_data_unb,fem_unb,mas_unb,
                        'Matriz de Correlação de Ambos os Genêros (Desbal.)',
                        'Matriz de Correlação do Genêro Feminino (Desbal.)',
                        'Matriz de Correlação do Genêro Masculino (Desbal.)',
                        "RdPu_r",'dengue_correlation_matrix_pt_unbalanced.png', 'pt')
```
